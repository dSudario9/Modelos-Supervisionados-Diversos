```python
# import da Biblioteca 
import yfinance as yf
```

EDA 

```python
# Importando dados das açoes WEGE3 01-01-2017  a 01-07-2021. 

# Esse modelo pode ser usado para extrair dados de qualquer outro papel . Se for ações brasileira ultilizar SA ao final das ações. 

dados= yf.download('WEGE3.SA',start="2007-01-01" , end= "2021-07-01")
```

```python
# Vizualizar dados 
dados.head()
```

```python
# Renomeando Colunas 
dados.rename(columns={'Open': 'Abrir','High': 'Alto','Low': 'Baixo','Close': 'Fechar','Adj Close': 'Adj Fechar'},inplace = True)
```
